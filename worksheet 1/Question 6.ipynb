```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

papers = soup.find_all('div', {'class': 'article-listing'})
papersList=[]
# print(papers[0])
for paper in papers:
    title = paper.find('a').text.strip()
    author = paper.find('p', {'class': 'article-authors'}).text
    date = paper.find('p', {'class': 'article-date'}).text
    papersList.append([title,author,date])
    

df = pd.DataFrame(papersList, columns=['Title', 'Author', 'Date'])
df
```
