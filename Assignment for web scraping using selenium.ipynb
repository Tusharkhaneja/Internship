```python
driver.get('https://www.flipkart.com')
```

```python
searchInput = driver.find_element(By.CLASS_NAME,'Pke_EE')
searchInput.send_keys('Sneakers')
```

```python
searchButton = driver.find_element(By.XPATH,'//button[@type="submit"]')
searchButton.click()
```

```python
brands=[]
decriptions=[]
prices=[]
```

```python
for i in range(0,3):
    brand_divs = driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    
    for j in brand_divs:
        brand = j.text
        brands.append(brand)
    
    description_divs = driver.find_elements(By.XPATH,'//a[@class="WKTcLC"]')
    
    for j in description_divs:
        if(len(decriptions) <= 100):
            decription = j.text
            decriptions.append(decription)
    
    price_divs = driver.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')
    
    for j in price_divs:
        price = j.text
        prices.append(price)

    if(i==0):
        next_Button = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
        next_Button.click()
    else:
        next_Button = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[12]')
        next_Button.click()

    time.sleep(5)
```

```python
print(len(brands),len(decriptions),len(prices))
del brands[len(decriptions):len(brands)]
del prices[len(decriptions):len(prices)]
print(len(brands),len(decriptions),len(prices))
```

```python
df=pd.DataFrame({'Brands':brands,'Decriptions':decriptions,'Prices':prices})
df
```

```python
for i in range(0,10):
    div_tags = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')

    for j in div_tags:
            Quote = j.text
            Quotes.append(Quote)
    
    Author_tags = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/div[1]/a')
    
    for j in Author_tags:
            Author = j.text
            Authors.append(Author)
    
    Type_tags = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/div[2]/div')
    
    for j in Type_tags:
            Type = j.text
            Types.append(Type)

    if(i==0):
        next_Button = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
        next_Button.click()
    else:
        next_Button = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
        next_Button.click()

    time.sleep(5)
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1')
```

```python
names = []
termOfOffices = []
Durations = []
AgeStarted = []
```

```python
for i in range(1,20):
    if(i != 1):
        brand_div = driver.find_element(By.XPATH,'//div[@class="TableData"]/table/tbody/tr['+ str(i) +']/td[2]')
        name = brand_div.text
        names.append(name)

        termOfOffice_div = driver.find_element(By.XPATH,'//div[@class="TableData"]/table/tbody/tr['+ str(i) +']/td[4]')
        termOfOffice_div2 = driver.find_element(By.XPATH,'//div[@class="TableData"]/table/tbody/tr['+ str(i) +']/td[5]')
        termOfOffice = termOfOffice_div.text + ' - ' + termOfOffice_div2.text
        termOfOffices.append(termOfOffice)

        Duration_div = driver.find_element(By.XPATH,'//div[@class="TableData"]/table/tbody/tr['+ str(i) +']/td[6]')
        Duration = Duration_div.text
        Durations.append(Duration)

        AgeStarted_div = driver.find_element(By.XPATH,'//div[@class="TableData"]/table/tbody/tr['+ str(i) +']/td[3]')
        Age = AgeStarted_div.text
        AgeStarted.append(Age)

print(AgeStarted)
```

```python
df=pd.DataFrame({'Names':names,'Term Of Offices':termOfOffices,'Durations':Durations,'Age Started':AgeStarted})
df
```

```python
names=[]
prices=[]
details=[]
```

```python
for i in range(1,51):
    name_tag = driver.find_element(By.XPATH,'//div[@class="postBody description e-content"]/h3['+ str(i) +']')
    name = name_tag.text
    names.append(name)
    
for i in range(4,104):
    if(i%2 == 0):
        price_tag = driver.find_element(By.XPATH,'//div[@class="postBody description e-content"]/p['+ str(i) +']/strong')
        price = price_tag.text
        prices.append(price)

for i in range(4,104):
    if(i%2 == 1):
        detail_tag = driver.find_element(By.XPATH,'//div[@class="postBody description e-content"]/p['+ str(i) +']')
        detail = detail_tag.text
        details.append(detail)

print(len(names),len(prices),len(details))
print(details)
```

```python
df=pd.DataFrame({'Names':names,'Prices':prices,'Details':details})
df
```
